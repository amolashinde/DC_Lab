```python
import Pyro4

@Pyro4.expose
class StringConcatenator:
    def concatenate(self, str1, str2):
        return str1 + str2

def start_server():
    daemon = Pyro4.Daemon()
    obj = StringConcatenator()

    url = daemon.register(obj)

    print("Server URL:", url)
    print("Server is ready...")

    daemon.requestLoop()

if __name__ == "__main__":
    start_server()
```
